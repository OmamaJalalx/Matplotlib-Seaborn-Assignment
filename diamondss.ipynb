```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("diamonds")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['price'], kde=True)
plt.title("Distribution of prices")
plt.show()
```

```python
sns.boxplot(x='cut',y='price',data=df)
plt.title("price distribution by cut")
plt.show()
```

```python
sns.scatterplot(x='carat', y='price',hue='color',data=df) 
plt.title("Price vs carat by color")
plt.show()
```

```python
sns.boxplot(x='color', y='price',data=df)
plt.title(" Price by Color")
plt.show()
```

```python
numericaldf= df.select_dtypes(include=['float64', 'int64'])
corr = numericaldf.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Replace x,y,z dimensions with size 

```python
df['x'] = pd.to_numeric(df['x'],errors='coerce')
df['y'] = pd.to_numeric(df['y'],errors='coerce')
df['z'] = pd.to_numeric(df['z'],errors='coerce')
df['size'] = df['x'] * df['y'] * df['z']
df['size']
```

```python
numericdf=df.select_dtypes(include=['float64', 'int64'])
numericdf.drop(columns=['x','y','z'], inplace=True)
corr=numericdf.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(x="size",y='price',data=df)
plt.title("price vs. size")
plt.show()
```

```python
sns.violinplot(x="clarity",y='price',data=df)
plt.title("price distribution by clarity")
plt.show()
```
